```python
#Importing Libraries
import requests
from bs4 import BeautifulSoup as bs
import csv
import pandas as pd
```

```python
 # url
url = "https://www.jobat.be/fr/emplois/results?k=d%c3%a9veloppeur"
```

```python

# Creating the request and veiwing the source of the page

page= requests.get(url)

page.close()
src= page.content
src
```

```python
# parsing the source
soup = bs(src, "lxml")
soup
```

```python
# Accessing Page Elements

jobresult = soup.find_all("div",{"class":"jobResults-card"})
jobcard = jobresult[0] # accessing the first element
job_title = jobcard.find('h2',{'class':'jobTitle'}).text.strip()
company = jobcard.find('li',{'class':'jobCard-company'}).text.strip()
web_url = jobcard.find('h2',{'class':'jobTitle'}).a.get("href")
job_url = "https://www.jobat.be"+str(web_url)
print(job_url)
location = jobcard.find('li',{'class':'jobCard-location'}).text.strip()
print(f"Job title: {job_title} \nCompany: {company} \nLocation: {location}")
```

```python
# Bringing it all together
jobcards = []
for i in range(len(jobresult)):
    jobcard = jobresult[i]
    
    job_title = jobcard.find('h2',{'class':'jobTitle'}).text.strip()
    company = jobcard.find('li',{'class':'jobCard-company'}).text.strip()
    location = jobcard.find('li',{'class':'jobCard-location'}).text.strip()
    web_url = jobcard.find('h2',{'class':'jobTitle'}).a.get("href")
    job_url = "https://www.jobat.be"+str(web_url)

    jobcards.append({"job_title":job_title,"company":company,"location":location,"url":job_url})
```

```python
jobcards
jobb = pd.DataFrame(jobcards)
```

```python
# save the data into excel file
jobb.to_excel("job_results.xlsx",index = False)
```

```python
# or creating a csv file to save our data into it
keys = jobcards[0].keys()
with open("job_results.csv",'w',encoding='utf-8') as my_file:
    dict_writer = csv.DictWriter(my_file,keys)
    dict_writer.writeheader()
    dict_writer.writerows(jobcards)
    print('file done')
```
